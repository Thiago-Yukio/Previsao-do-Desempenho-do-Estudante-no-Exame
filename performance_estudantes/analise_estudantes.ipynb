```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier,GradientBoostingClassifier,HistGradientBoostingClassifier
from sklearn.tree import DecisionTreeRegressor,DecisionTreeClassifier
```

```python
def classificar_nota(nota):
    if nota > 75.0:
        return 2
    elif nota > 50.0:
        return 1
    else:
        return 0
df['classificacao_exame'] = df['nota_exame'].apply(classificar_nota)
df['classificacao_exame'].value_counts()
```

```python
features_modelo=['idade', 'horas_estudo_dia',
       'horas_rede_social', 'horas_netflix', 'trabalho_meio_período',
       'porcentagem_de_presenca', 'horas_sono',
       'frequencia_exercicios', 'avaliacao_saude_mental','participacao_atividades_extracurriculares']
alvo_modelo=['classificacao_exame']
```

```python
df_modelo=df[features_modelo+alvo_modelo].copy()
df_modelo.info()
```

```python
encoder=LabelEncoder()
df_modelo["trabalho_meio_período"] = encoder.fit_transform(df_modelo["trabalho_meio_período"])
df_modelo['participacao_atividades_extracurriculares']=encoder.fit_transform(df_modelo['participacao_atividades_extracurriculares'])
#df_modelo['qualidade_internet']=encoder.fit_transform(df_modelo['qualidade_internet'])
```

```python
df_modelo.info()
```

```python
X=df_modelo[features_modelo]
y=df_modelo[alvo_modelo]
```

```python
X_treino,X_teste,y_treino,y_teste=train_test_split(X,y,test_size=0.3)
```

```python
modelos = {
    "DecisionTreeClassifier": {
        "modelo": DecisionTreeClassifier(class_weight='balanced',random_state=42),
        "parametros": {"max_depth": [3,5,10], "min_samples_split": [2,5]}
    },
        "RandomForestRegressor" : {
        "modelo": RandomForestRegressor(random_state=42),
        "parametros": {'n_estimators': [400], 'min_samples_split': [2], 'min_samples_leaf': [1], 'max_features': [1.0], 'max_depth': [20]}
    },
    "DecisionTreeRegressor" : {
        "modelo" : DecisionTreeRegressor(random_state=42),
        "parametros": {"max_depth": [3,5,10], "min_samples_split": [2,5]}
    },
        "RandomForestClassifier" : {
        "modelo": RandomForestClassifier(class_weight='balanced',random_state=42),
        "parametros": {"n_estimators": [ 50,100], "max_depth": [5,10]}
    },
        "GradientBoostingClassifier": {
        "modelo": GradientBoostingClassifier(random_state=42),
        "parametros": {"max_depth": [3,5,10], "min_samples_split": [2,5]}
    },
        "HistGradientBoostingClassifier ": {
        "modelo": HistGradientBoostingClassifier (class_weight='balanced',max_depth=3,random_state=42),
        "parametros": {'l2_regularization': [1.0],'learning_rate': [0.05],'max_iter': [300],'max_leaf_nodes': [31] }
}}
for i in modelos:
    print(i)
```

```python
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import (
    mean_squared_error, r2_score,
    accuracy_score, f1_score
)

melhor_modelos = []

for name, config in modelos.items():
    print(f"Treinando {name}...")

    is_classifier = "Classifier" in name

    scoring_metric = "accuracy" if is_classifier else "neg_mean_squared_error"

    grid = GridSearchCV(
        estimator=config["modelo"],
        param_grid=config["parametros"],
        cv=5,
        scoring=scoring_metric,
        n_jobs=-1
    )
    grid.fit(X_treino, y_treino)

    y_pred = grid.predict(X_teste)

    resultado = {
        "modelo": name,
        "best_params": grid.best_params_
    }

    if is_classifier:
        resultado["acuracia"] = accuracy_score(y_teste, y_pred)
        resultado["f1_macro"] = f1_score(y_teste, y_pred, average="macro")
        resultado["tipo"] = "Classificação"
    else:
        resultado["rmse"] = np.sqrt(mean_squared_error(y_teste, y_pred))
        resultado["r2"] = r2_score(y_teste, y_pred)
        resultado["tipo"] = "Regressão"

    melhor_modelos.append(resultado)

print("\n--- Treinamento Concluído! ---")
```

```python
resulatado_df=pd.DataFrame(melhor_modelos)
resulatado_df
```
